```python
from emb_diversity import plot_2d, plot_3d
import numpy as np
import warnings
```

```python
vectors_1 = np.random.randn(100, 384) * 5.0 # high variance
vectors_2 = np.random.randn(100, 384) * 0.5 # low variance
vectors_1
```

# Two-dimensional Plotting

### We support plotting with t-SNE, UMAP and PCA (default):

```python
plot_2d(vectors_1, method="tsne")
```

```python
plot_2d(vectors_1, method="umap")
```

```python
plot_2d(vectors_1, method="pca")
```

### We support plotting multiple datasets simultaneously:

```python
plot_2d([vectors_1, vectors_2], method="tsne")
```

```python
# Dataset labels can be given manually
plot_2d([vectors_1, vectors_2], method="pca", labels=["d1", "d2"])
```

# Three-dimensional Plotting

```python
plot_3d([vectors_1, vectors_2], method="umap")
```

```python
plot_3d([vectors_1, vectors_2], method="tsne", labels=["Dataset A", "Dataset B"])
```

```python
plot_3d([vectors_1, vectors_2], method="pca")
```
