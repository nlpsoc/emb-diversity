```python
import re
def extract_formal_informal_pairs(tsv_file_path):
    """
    Extract unique formal-informal sentence pairs from a TSV file.

    Args:
        tsv_file_path (str): Path to the TSV file

    Returns:
        pd.DataFrame: DataFrame with 'formal' and 'informal' columns containing unique pairs
    """
    # Read the TSV file
    df = pd.read_csv(tsv_file_path, sep='\t')

    # Dictionary to store sentences by their unique ID
    sentences_by_id = {}

    # Process each row
    for _, row in df.iterrows():
        id_field = row['ID']

        # Extract all unique identifiers from the ID field
        # Pattern matches i-XXXX or f-XXXX
        matches = re.findall(r'([if])-(\d+)', id_field)

        # Get the sentence columns (excluding metadata columns)
        sentence_columns = ['Anchor 1', 'Anchor 2', 'Alternative 1.1', 'Alternative 1.2']

        # Map each sentence to its formality type based on position and ID pattern
        for i, match in enumerate(matches):
            formality_type, unique_id = match

            # Get the corresponding sentence (assumes order matches)
            if i < len(sentence_columns):
                sentence = row[sentence_columns[i]]

                # Store in dictionary by unique ID
                if unique_id not in sentences_by_id:
                    sentences_by_id[unique_id] = {}

                sentences_by_id[unique_id][formality_type] = sentence

    # Create pairs from sentences that have both formal and informal versions
    formal_sentences = []
    informal_sentences = []

    for unique_id, sentences in sentences_by_id.items():
        if 'f' in sentences and 'i' in sentences:
            formal_sentences.append(sentences['f'])
            informal_sentences.append(sentences['i'])

    # Create DataFrame with unique pairs
    pairs_df = pd.DataFrame({
        'formal': formal_sentences,
        'informal': informal_sentences
    })

    # Remove duplicate pairs
    pairs_df = pairs_df.drop_duplicates().reset_index(drop=True)

    return pairs_df
```
