```python
unique_values = set(dataset['train']["feature"])
unique_values
```

```python
# Create a dictionary to store examples for each unique feature
seen_features = set()
examples = {}

for item in dataset['train']:
    feature_value = item['feature']
    if feature_value not in seen_features:
        seen_features.add(feature_value)
        examples[feature_value] = item
```
